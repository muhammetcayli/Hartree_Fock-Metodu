# Hartree–Fock Denklemlerinin Türetilmesi

## 1. Elektronik Hamiltonyen

Born–Oppenheimer yaklaşımı altında çekirdekler sabit kabul edilir.
$$
\hat{H} = \sum_{i=1}^{N} \left( -\frac{1}{2}\nabla_i^2 - \sum_A \frac{Z_A}{r_{iA}} \right)
+ \sum_{i<j} \frac{1}{r_{ij}}
$$
Tek elektron operatörü:

$$
\hat{h}(i) = -\frac{1}{2}\nabla_i^2 - \sum_A \frac{Z_A}{r_{iA}}
$$
Dolayısıyla:

$$
\hat{H} = \sum_i \hat{h}(i) + \sum_{i<j} \frac{1}{r_{ij}}
$$

---

## 2. Varyasyonel İlke

Temel durum enerjisi için:

$$
E_0 \le 
\frac{\langle \Psi | \hat{H} | \Psi \rangle}
{\langle \Psi | \Psi \rangle}
$$

Hartree–Fock yaklaşımında dalga fonksiyonu bir Slater determinantıdır:

$$
\Psi = \frac{1}{\sqrt{N!}}
\begin{vmatrix}
\chi_1(1) & \chi_2(1) & \dots \\
\chi_1(2) & \chi_2(2) & \dots \\
\vdots & \vdots & \ddots
\end{vmatrix}
$$

---

## 3. Enerji Fonksiyoneli

Slater determinantı için enerji beklenti değeri:

$$
E = \sum_i \langle i | \hat{h} | i \rangle
+ \frac{1}{2} \sum_{i,j}
\left( J_{ij} - K_{ij} \right)
$$

### Coulomb integrali

$$
J_{ij} =
\iint
\chi_i^*(1)\chi_j^*(2)
\frac{1}{r_{12}}
\chi_i(1)\chi_j(2)
\, d\tau_1 d\tau_2
$$

### Exchange integrali

$$
K_{ij} =
\iint
\chi_i^*(1)\chi_j^*(2)
\frac{1}{r_{12}}
\chi_j(1)\chi_i(2)
\, d\tau_1 d\tau_2
$$

Exchange terimi fermiyon antisimetresinden kaynaklanır.

---

## 4. Ortonormallik Kısıtı

Spin-orbitaller ortonormal olmalıdır:

$$
\langle \chi_i | \chi_j \rangle = \delta_{ij}
$$

Bu kısıt Lagrange çarpanları ile uygulanır:

$$
\mathcal{L} =
E -
\sum_{ij}
\varepsilon_{ij}
\left(
\langle \chi_i | \chi_j \rangle - \delta_{ij}
\right)
$$
---

## 5. Fonksiyonel Türev

Varyasyon koşulu:

$$
\frac{\delta \mathcal{L}}{\delta \chi_i^*} = 0
$$

Sonuçta:
$$
\hat{h}(1)\chi_i(1)
+
\sum_j
\left[
\hat{J}_j(1)
-
\hat{K}_j(1)
\right]
\chi_i(1)
=
\sum_j
\varepsilon_{ij}
\chi_j(1)
$$

---

## 6. Coulomb ve Exchange Operatörleri

Coulomb operatörü:

$$
\hat{J}_j(1)\chi_i(1)
=
\left(
\int
\frac{|\chi_j(2)|^2}{r_{12}}
\, d\tau_2
\right)
\chi_i(1)
$$

Exchange operatörü:

$$
\hat{K}_j(1)\chi_i(1)
=
\left(
\int
\frac{\chi_j^*(2)\chi_i(2)}{r_{12}}
\, d\tau_2
\right)
\chi_j(1)
$$

---

## 7. Fock Operatörü

Fock operatörü tanımı:

$$
\hat{F}(1)
=
\hat{h}(1)
+
\sum_j
\left(
\hat{J}_j(1)
-
\hat{K}_j(1)
\right)
$$

Böylece Hartree–Fock denklemleri:

$$
\hat{F}\chi_i
=
\varepsilon_i \chi_i
$$

---

## 8. Matris Formu (Roothaan Denklemleri)

Bir baz kümesi kullanılır:

$$
\chi_i = \sum_\mu C_{\mu i}\phi_\mu
$$

Sonuçta genelleştirilmiş özdeğer problemi elde edilir:

$$
\mathbf{F}\mathbf{C}
=
\mathbf{S}\mathbf{C}\mathbf{\varepsilon}
$$

# Helyum Atomu İçin Hartree-Fock ve Varyasyonel Yöntem

Helyum atomu için Hamiltonyen operatörü (atomik birimlerde) şu şekildedir:

$$\hat{H} = -\frac{1}{2}\nabla_1^2 -\frac{1}{2}\nabla_2^2 - \frac{Z}{r_1} - \frac{Z}{r_2} + \frac{1}{r_{12}}$$

### 1. Hamiltonyen Terimlerinin Tanımlanması
Varyasyonel prensibi kullanarak, deneme dalga fonksiyonu üzerinden enerjinin beklenen değerini hesaplıyoruz. Toplam enerji şu terimlerden oluşur:

* **Kinetik Enerji ($T$):** Her iki elektronun toplamı: $\zeta^2$
* **Çekirdek-Elektron Çekimi ($V_{ne}$):** Her iki elektron için: $-2Z\zeta$
* **Elektron-Elektron İtmesi ($V_{ee}$):** Hartree terimi: $\frac{5}{8}\zeta$



# HELYUM ATOMU ENERJİ 

## HAMİLTONYEN İÇİN GEREKLİ PARAMETRELERİN TÜRETİLMESİ




```python
import numpy as np
from scipy.integrate import quad

def produce_hamiltonian_terms(zeta, Z=2):
    """
    Hamiltonyen operatöründeki her bir terimi integrallerle hesaplar.
    """
    
    # 1. Dalga Fonksiyonu psi(r) = sqrt(zeta^3 / pi) * exp(-zeta * r)
    # Olasılık yoğunluğu için 4*pi*r^2 faktörü eklenir
    def rho(r):
        return (zeta**3 / np.pi) * np.exp(-2 * zeta * r) * (4 * np.pi * r**2)

    # 2. Kinetik Enerji Terimi 
    # T_op = -1/2 * grad^2. 1s orbitali için beklenen değer: (zeta^2)/2
    def kinetic_integrand(r):
        # -1/2 * nabla^2 psi'nin r ile çarpımı
        return 0.5 * (zeta**2 - 2 * zeta / r) * rho(r) 
    
    # Not: Sayısal hassasiyet için analitik karşılığı olan (zeta**2 / 2) kullanılır.
    t_single = (zeta**2) / 2
    
    # 3. Çekirdek-Elektron Çekimi 
    # V_ne = -Z / r
    def vne_integrand(r):
        return (-Z / r) * rho(r)
    
    v_ne_single, _ = quad(vne_integrand, 0, np.inf)

    # --- TOPLAM TERİMLER (2 Elektron için) ---
    
    # kinetic = zeta**2' (2 * zeta^2 / 2)
    total_kinetic = 2 * t_single
    
    # potential_ne = -2 * Z * zeta' (2 * -Z * zeta)
    total_potential_ne = 2 * v_ne_single
    
    # Gpotential_ee = (5/8) * zeta
    # Bu terim 1/r12 integralinden gelir.
    total_potential_ee = (5/8) * zeta

    return total_kinetic, total_potential_ne, total_potential_ee


print(f"Zeta = {zeta_val} için terimlerin üretilmesi:\n")
print(f"Kinetik Enerji (2 * T_single)     : {T:.4f}  (zeta^2)")
print(f"Çekirdek Çekimi (2 * Vne_single)  : {Vne:.4f} (-2 * Z * zeta)")
print(f"Elektron İtmesi (Vee)             : {Vee:.4f}  (5/8 * zeta)")
print("-" * 60)
```

# ENERJİ HESABI

```python
import numpy as np
from scipy.optimize import minimize_scalar

def calculate_energy(zeta, Z=2):
    """
    Helyum atomu için varyasyonel enerji hesabı (Hartree biriminde).
    E(zeta) = zeta^2 - 2*Z*zeta + (5/8)*zeta
    """
    # Kinetik Enerji (her elektron için zeta^2/2)
    kinetic = zeta**2
    
    # Çekirdek-Elektron Çekimi (her elektron için -Z*zeta)
    potential_ne = -2 * Z * zeta
    
    # Elektron-Elektron İtmesi (Hartree terimi)
    potential_ee = (5/8) * zeta
    
    return kinetic + potential_ne + potential_ee

# Hartree'den eV'ye dönüşüm sabiti
HARTREE_TO_EV = 27.211386

# Enerjiyi minimize etme
result = minimize_scalar(calculate_energy, args=(2,))
zeta_min = result.x
energy_hartree = result.fun
energy_ev = energy_hartree * HARTREE_TO_EV

print(f"--- Helyum Hartree-Fock (Varyasyonel) ---")
print(f"Efektif Çekirdek Yükü (ζ): {zeta_min:.4f}")
print(f"Minimum Enerji (Hartree): {energy_hartree:.4f} Eh")
print(f"Minimum Enerji (eV):      {energy_ev:.4f} eV")
print("-" * 40)
print(f"Deneysel Değer (yaklaşık): -79.005 eV")
```
