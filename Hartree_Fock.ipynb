# Hartree–Fock Denklemlerinin Türetilmesi

## 1. Elektronik Hamiltonyen

Born–Oppenheimer yaklaşımı altında çekirdekler sabit kabul edilir.
$$
\hat{H} = \sum_{i=1}^{N} \left( -\frac{1}{2}\nabla_i^2 - \sum_A \frac{Z_A}{r_{iA}} \right)
+ \sum_{i<j} \frac{1}{r_{ij}}
$$
Tek elektron operatörü:

$$
\hat{h}(i) = -\frac{1}{2}\nabla_i^2 - \sum_A \frac{Z_A}{r_{iA}}
$$
Dolayısıyla:

$$
\hat{H} = \sum_i \hat{h}(i) + \sum_{i<j} \frac{1}{r_{ij}}
$$

---

## 2. Varyasyonel İlke

Temel durum enerjisi için:

$$
E_0 \le 
\frac{\langle \Psi | \hat{H} | \Psi \rangle}
{\langle \Psi | \Psi \rangle}
$$

Hartree–Fock yaklaşımında dalga fonksiyonu bir Slater determinantıdır:

$$
\Psi = \frac{1}{\sqrt{N!}}
\begin{vmatrix}
\chi_1(1) & \chi_2(1) & \dots \\
\chi_1(2) & \chi_2(2) & \dots \\
\vdots & \vdots & \ddots
\end{vmatrix}
$$

---

## 3. Enerji Fonksiyoneli

Slater determinantı için enerji beklenti değeri:

$$
E = \sum_i \langle i | \hat{h} | i \rangle
+ \frac{1}{2} \sum_{i,j}
\left( J_{ij} - K_{ij} \right)
$$

### Coulomb integrali

$$
J_{ij} =
\iint
\chi_i^*(1)\chi_j^*(2)
\frac{1}{r_{12}}
\chi_i(1)\chi_j(2)
\, d\tau_1 d\tau_2
$$

### Exchange integrali

$$
K_{ij} =
\iint
\chi_i^*(1)\chi_j^*(2)
\frac{1}{r_{12}}
\chi_j(1)\chi_i(2)
\, d\tau_1 d\tau_2
$$

Exchange terimi fermiyon antisimetresinden kaynaklanır.

---

## 4. Ortonormallik Kısıtı

Spin-orbitaller ortonormal olmalıdır:

$$
\langle \chi_i | \chi_j \rangle = \delta_{ij}
$$

Bu kısıt Lagrange çarpanları ile uygulanır:

$$
\mathcal{L} =
E -
\sum_{ij}
\varepsilon_{ij}
\left(
\langle \chi_i | \chi_j \rangle - \delta_{ij}
\right)
$$
---

## 5. Fonksiyonel Türev

Varyasyon koşulu:

$$
\frac{\delta \mathcal{L}}{\delta \chi_i^*} = 0
$$

Sonuçta:
$$
\hat{h}(1)\chi_i(1)
+
\sum_j
\left[
\hat{J}_j(1)
-
\hat{K}_j(1)
\right]
\chi_i(1)
=
\sum_j
\varepsilon_{ij}
\chi_j(1)
$$

---

## 6. Coulomb ve Exchange Operatörleri

Coulomb operatörü:

$$
\hat{J}_j(1)\chi_i(1)
=
\left(
\int
\frac{|\chi_j(2)|^2}{r_{12}}
\, d\tau_2
\right)
\chi_i(1)
$$

Exchange operatörü:

$$
\hat{K}_j(1)\chi_i(1)
=
\left(
\int
\frac{\chi_j^*(2)\chi_i(2)}{r_{12}}
\, d\tau_2
\right)
\chi_j(1)
$$

---

## 7. Fock Operatörü

Fock operatörü tanımı:

$$
\hat{F}(1)
=
\hat{h}(1)
+
\sum_j
\left(
\hat{J}_j(1)
-
\hat{K}_j(1)
\right)
$$

Böylece Hartree–Fock denklemleri:

$$
\hat{F}\chi_i
=
\varepsilon_i \chi_i
$$

---

## 8. Matris Formu (Roothaan Denklemleri)

Bir baz kümesi kullanılır:

$$
\chi_i = \sum_\mu C_{\mu i}\phi_\mu
$$

Sonuçta genelleştirilmiş özdeğer problemi elde edilir:

$$
\mathbf{F}\mathbf{C}
=
\mathbf{S}\mathbf{C}\mathbf{\varepsilon}
$$

# HELYUM ATOMUNUN ENERJİSİ


## Hartree–Fock Yaklaşımı ile Helyum Enerjisi

- **Her iki elektron için hidrojen-benzeri 1s orbital:**

$$
\psi(r) = \frac{Z_{\mathrm{eff}}^{3/2}}{\sqrt{\pi}} e^{-Z_{\mathrm{eff}} r}
$$

---

- **Toplam enerji (Hartree biriminde):**

$$
E(Z_{\mathrm{eff}}) =
Z_{\mathrm{eff}}^2
- 2 Z Z_{\mathrm{eff}}
+ \frac{5}{8} Z_{\mathrm{eff}}
$$

---

### Burada:

- $Z = 2$ (Helyum için)
- Son terim elektron-elektron itmesidir
- $1$ Hartree = $27.2114$ eV


```python
import numpy as np
from scipy.optimize import minimize

# Sabitler
Z = 2  # Helyum çekirdek yükü
hartree_to_ev = 27.2114

# Enerji fonksiyonu (Hartree biriminde)
def helium_energy(Zeff):
    Zeff = Zeff[0]
    E = Zeff**2 - 2*Z*Zeff + (5/8)*Zeff
    return E

# Başlangıç tahmini
initial_guess = [1.7]

# Minimizasyon
result = minimize(helium_energy, initial_guess)

Zeff_opt = result.x[0]
E_opt_hartree = result.fun
E_opt_ev = E_opt_hartree * hartree_to_ev


print("Minimum Energy (Hartree) =", E_opt_hartree)
print("Minimum Energy (eV) =", E_opt_ev)
```
