# HELYUM ATOMU İÇİN HARTREE-FOCK METHODU





## Schrödinger Denklemi


$
H \psi = E \psi
$

$
H =
-\frac{1}{2}\nabla^2
+
V(r)
$


## Helium Çekirdek Potansiyeli


$
V_{nuc}(r) = -\frac{Z}{r}
$

$
Z = 2
$

## Radyal Dalga Fonksiyonu Tanımı


$
u(r) = rR(r)
$


## Hidrojen-benzeri 1s Orbitali


$s
R_{1s}(r) = 2Z^{3/2} e^{-Zr}
$

$
u(r) = 2r e^{-Zr}
$

## Normalize Koşulu


$
\int_0^\infty u(r)^2 dr = 1
$


## Elektron Yoğunluğu


$
\rho(r) = |\psi(r)|^2
$

$
\rho(r) = u(r)^2
$


## Hartree Potansiyeli


$
V_H(r)=
\frac{1}{r}
\int_0^r
\rho(r')dr'
+
\int_r^\infty
\frac{\rho(r')}{r'}dr'
$

## İç Yük


$
Q(r) =
\int_0^r
\rho(r')dr'
$


## Efektif Potansiyel


$
V_{eff}(r)
=
-\frac{Z}{r}
+
V_H(r)
$


## Radyal Schrödinger Denklemi


$
\left[
-\frac{1}{2}
\frac{d^2}{dr^2}
+
V_{eff}(r)
\right]
u(r)
=
\epsilon u(r)
$


## Sonlu Fark Yaklaşımı


$
\frac{d^2u}{dr^2}
\approx
\frac{u_{i+1}-2u_i+u_{i-1}}{dr^2}
$


## Hamiltonyen Matris Elemanları


$
H_{ii}
=
\frac{1}{dr^2}
+
V_{eff}(r_i)
$

$
H_{i,i\pm1}
=
-\frac{1}{2dr^2}
$


## Özdeğer Problemi


$
Hu = \epsilon u
$


## Hartree-Fock Toplam Enerjisi


s$
E =
2\epsilon
-
\int
V_H(r)\rho(r)dr
$


## Enerji Birim Dönüşümü


$
1 \, \text{Hartree}
=
27.211386 \, \text{eV}
$

```python
import numpy as np
from scipy.linalg import eigh_tridiagonal

HARTREE_TO_EV = 27.211386

# ---------- GRID ----------
Z = 2.0 #HELYUM ATOMUNUN ÇEKİRDEK YÜKÜ
r_max = 20.0
N = 2000

dr = r_max / N
r = np.linspace(dr, r_max, N)


u = 2 * r * np.exp(-Z * r) #HELYUM İÇİN BAŞLANGIÇ ORBİTALİ
u /= np.sqrt(np.trapezoid(u**2, r)) #DALGA FONKSİYONUNU NORMALİZASYONU

def hartree_potential(u): #Hartree-Fock’ta her elektron diğerinin oluşturduğu elektrostatik potansiyeli hisseder
    
    rho_radial = u**2  # YOĞUNLUK 
    
    # İçeride kalan yük (Q)
    Q = np.cumsum(rho_radial) * dr
    term1 = Q / r # İÇ KISIM İÇİM COULOMB POTANSİYELİ 

    
    term2 = np.cumsum((rho_radial / r)[::-1])[::-1] * dr #DIŞ KISIM İÇİN POTANSİYEL

    return term1 + term2 # İÇ VE DIŞ KISMIN OLUŞTURDUĞU TOPLAM POTANSİYEL

threshold = 1e-7 #SCF DÖNGÜSÜNÜN BAŞLANGICI
E_old = 0

for iteration in range(100):

    V_H = hartree_potential(u)
    V_eff = -Z/r + V_H

    
    diag = 1/dr**2 + V_eff #MATRİS ELEMANLARI
    off = -0.5/dr**2 * np.ones(N-1)

    epsilon, vecs = eigh_tridiagonal(diag, off) #ÖZDEĞER PROBLEMİ

    u_new = vecs[:, 0] #1S ORBİTALİ YANİ EN DÜŞÜK ENERJİLİ ORBİTALİ SEÇİYORUZ
    u_new /= np.sqrt(np.trapezoid(u_new**2, r)) #YENİ DALGA FONKSİYONUNU NORMALİZASYONU


    E_total = 2 * epsilon[0] - np.trapezoid(V_H * (u_new**2), r) #HARTREE-FOCK TOPLAM ENERİSİ

    print(f"{iteration+1:2d}. Iterasyon: {E_total * HARTREE_TO_EV:.4f} eV")

    if abs(E_total - E_old) < threshold:
        break

    u = u_new
    E_old = E_total

print("\nFinal HF Energy (Hartree):", E_total)
print("Final HF Energy (eV):", E_total * HARTREE_TO_EV)
```
