

# Hartree–Fock Yöntemi 

---

# 1. Problemin Tanımı

$N$ elektronlu bir sistem için Hamiltonyen:

$$
\hat{H} = \sum_{i=1}^{N} \hat{h}(i) + \sum_{i<j}^{N} \frac{1}{r_{ij}}
$$

Burada:

$$
\hat{h}(i) = -\frac{1}{2}\nabla_i^2 + V_{ext}(\mathbf{r}_i)
$$

Amaç:

$$
\hat{H} \Psi = E \Psi
$$

denklemini çözmek.

---

# 2. Varyasyonel Prensip

Temel prensip:

$$
E_0 \leq \frac{\langle \Psi | \hat{H} | \Psi \rangle}{\langle \Psi | \Psi \rangle}
$$

Dolayısıyla:

$$
\delta \langle \Psi | \hat{H} | \Psi \rangle = 0
$$

---

# 3. Antisimetri ve Slater Determinantı

Fermiyonlar için:

$$
\Psi(...,x_i,...,x_j,...) = -\Psi(...,x_j,...,x_i,...)
$$

Bu koşulu sağlayan en basit yapı:

$$
\Psi = \frac{1}{\sqrt{N!}} \det[\phi_i(x_j)]
$$

---

# 4. Enerji Fonksiyonelinin Açılması

Enerji:

$$
E = \langle \Psi | \hat{H} | \Psi \rangle
$$

Determinant kullanılarak açıldığında iki tür integral çıkar:

---

## 4.1 Tek Parçacık Terimi

$$
\sum_i \int \phi_i^*(1) \hat{h}(1) \phi_i(1) d\tau_1
$$

---

## 4.2 İki Parçacık Terimi

Determinant genişletilince:

$$
\frac{1}{2} \sum_{ij} \left( \langle ij | \frac{1}{r_{12}} | ij \rangle - \langle ij | \frac{1}{r_{12}} | ji \rangle \right)
$$

---

# 5. Coulomb ve Exchange Terimlerinin Türetilmesi

## 5.1 Coulomb Terimi

$$
J_{ij} = \int \int \phi_i^*(1)\phi_j^*(2)\frac{1}{r_{12}}\phi_i(1)\phi_j(2) d\tau_1 d\tau_2
$$

Bu terim klasik elektrostatik enerjiye karşılık gelir.

---

## 5.2 Exchange Terimi

$$
K_{ij} = \int \int \phi_i^*(1)\phi_j^*(2)\frac{1}{r_{12}}\phi_j(1)\phi_i(2) d\tau_1 d\tau_2
$$

Bu tamamen antisimetri sonucu ortaya çıkar.

---

# 6. Enerji Fonksiyonelinin Tam Formu

$$
E[\{\phi_i\}] =
\sum_i \langle i | h | i \rangle
+ \frac{1}{2} \sum_{ij} (J_{ij} - K_{ij})
$$

---

# 7. Kısıtlı Varyasyon (Lagrange Çarpanları)

Ortonormallik:

$$
\langle \phi_i | \phi_j \rangle = \delta_{ij}
$$

Fonksiyonel:

$$
\mathcal{L} =
E - \sum_{ij} \epsilon_{ij} (\langle \phi_i | \phi_j \rangle - \delta_{ij})
$$

---

# 8. Fonksiyonel Türev

$$
\frac{\delta \mathcal{L}}{\delta \phi_i^*} = 0
$$

Her terimi ayrı türevleyelim:

---

## 8.1 Tek Parçacık Terimi

$$
\delta \langle i | h | i \rangle \rightarrow \hat{h}\phi_i
$$

---

## 8.2 Coulomb Terimi

$$
\delta J_{ij} \rightarrow \left( \int \frac{|\phi_j(2)|^2}{r_{12}} d\tau_2 \right)\phi_i(1)
$$

---

## 8.3 Exchange Terimi

$$
\delta K_{ij} \rightarrow
\left( \int \phi_j^*(2)\frac{1}{r_{12}}\phi_i(2)d\tau_2 \right)\phi_j(1)
$$

---

# 9. Fock Operatörünün Ortaya Çıkışı

Tüm terimler birleştirilirse:

$$
\hat{F}\phi_i = \epsilon_i \phi_i
$$

Burada:

$$
\hat{F} = \hat{h} + \sum_j (\hat{J}_j - \hat{K}_j)
$$

---

# 10. Bu Denklemin Anlamı

Bu aslında:

- Çok parçacıklı problem → tek parçacıklı denklemler
- Ama operatör orbitallere bağlı → doğrusal değil

---

# 11. Self-Consistent Field (SCF)

# 1. Problemin Doğası: Neden SCF Gerekli?

Hartree–Fock denklemi:

$$
\hat{F}[\{\phi_j\}] \phi_i = \epsilon_i \phi_i
$$

Burada kritik nokta:

$$
\hat{F} = \hat{h} + \sum_j (\hat{J}_j - \hat{K}_j)
$$

Ama:

$$
\hat{J}_j, \hat{K}_j \propto \phi_j
$$

Dolayısıyla:

$$
\hat{F} = \hat{F}[\phi]
$$

Bu şu anlama gelir:

> Bu bir **nonlineer özdeğer problemidir**

---

# 2. SCF Fikrinin Temeli

Amaç:

$$
\phi_i^{(n)} \rightarrow \phi_i^{(n+1)}
$$

şeklinde iteratif bir süreç kurmak.

---

# 3. Başlangıç Yaklaşımı

İlk orbitaller:

$$
\{\phi_i^{(0)}\}
$$

Genelde:

- Atomik orbitaller
- Minimal basis
- Core Hamiltonian çözümü

---

# 4. Yoğunluk (Density) Kavramı

SCF’nin kalbi:

$$
\rho(\mathbf{r}) = \sum_i |\phi_i(\mathbf{r})|^2
$$

Matris formunda:

$$
P_{\mu\nu} = \sum_i C_{\mu i} C_{\nu i}
$$

---

# 5. Fock Operatörünün İnşası

## 5.1 Tek Parçacık Terimi

$$
h_{\mu\nu} = \int \chi_\mu^*(1) \hat{h}(1) \chi_\nu(1) d\tau
$$

---

## 5.2 İki Elektron Integralleri

$$
(\mu\nu|\lambda\sigma) =
\int\int \chi_\mu(1)\chi_\nu(1)\frac{1}{r_{12}}\chi_\lambda(2)\chi_\sigma(2) d\tau_1 d\tau_2
$$

---

## 5.3 Fock Matrisi

$$
F_{\mu\nu} = h_{\mu\nu} + \sum_{\lambda\sigma} P_{\lambda\sigma}
\left[
(\mu\nu|\lambda\sigma) - \frac{1}{2}(\mu\sigma|\lambda\nu)
\right]
$$

---

# 6. Roothaan Denklemleri

$$
\mathbf{F} \mathbf{C} = \mathbf{S} \mathbf{C} \mathbf{\epsilon}
$$

Burada:

- $\mathbf{S}$ → overlap matrisi
- $\mathbf{C}$ → orbital katsayıları

---

# 7. Ortogonalizasyon

Genelleştirilmiş özdeğer problemi:

$$
\mathbf{S}^{-1/2} \mathbf{F} \mathbf{S}^{-1/2} \mathbf{C}' = \mathbf{C}' \epsilon
$$

---

# 8. SCF Döngüsü (Matematiksel Akış)

## Iterasyon:

### 1. Yoğunluk matrisi oluştur:

$$
P^{(n)}_{\mu\nu} = \sum_i C^{(n)}_{\mu i} C^{(n)}_{\nu i}
$$

---

### 2. Fock matrisi oluştur:

$$
F^{(n)} = h + G[P^{(n)}]
$$

---

### 3. Özdeğer problemini çöz:

$$
F^{(n)} C^{(n+1)} = S C^{(n+1)} \epsilon
$$

---

### 4. Yeni yoğunluk:

$$
P^{(n+1)} = C^{(n+1)} C^{(n+1)T}
$$

---

### 5. Enerji hesapla:

$$
E^{(n)} =
\sum_{\mu\nu} P_{\mu\nu}(h_{\mu\nu} + F_{\mu\nu})
$$

---

### 6. Yakınsama kontrolü:

$$
||P^{(n+1)} - P^{(n)}|| < \epsilon
$$

---

### 9. Yakınsama Problemi

SCF her zaman yakınsamaz!

Sebep:

- Nonlineer yapı
- Exchange terimi

---

### 10. Kararlılık Analizi

HF aslında şu fonksiyoneli minimize eder:

$$
E[P]
$$

SCF → sabit nokta iterasyonu:

$$
P^{(n+1)} = \mathcal{F}(P^{(n)})
$$

---

### 11. İyileştirme Teknikleri

## 11.1 Damping

$$
P^{(n+1)} = \alpha P^{(n+1)} + (1-\alpha)P^{(n)}
$$

---

## 11.2 DIIS (Direct Inversion)

Hata vektörü:

$$
R = FP - PF
$$

Amaç:

$$
R \rightarrow 0
$$

---

### 12. Fiziksel Yorum

SCF şu anlama gelir:

> Elektronlar kendi oluşturdukları ortalama potansiyelde hareket eder

---

### 13. Kritik İçgörü

SCF aslında:

$$
\text{Fixed-point iteration}
$$

şeklinde bir problemdir.

---

### 14. Sonuç

SCF döngüsü:

- Nonlineer problemi lineer adımlara böler
- Her adımda yeni "evren" oluşturur
- Kendi kendine tutarlı çözüm bulur

---

### SON CÜMLE

Hartree–Fock çözümü:

$$
\boxed{
\text{Self-consistent orbitals}
}
$$

yani:

> Orbitaller kendi oluşturdukları alanla tutarlı hale gelene kadar iterasyon yapılır.


# 12. Fiziksel Yorum

- $J$ → klasik itme
- $K$ → Pauli prensibi etkisi
- HF → ortalama alan teorisi

---

# 13. Önemli Sonuç

Hartree–Fock:

- Dalga fonksiyonunu determinant ile kurar
- Enerjiyi minimize eder
- Exchange etkisini içerir
- Ama korelasyon eksiktir

---

# 14. Kritik Nokta (En Önemli İçgörü)

HF denklemleri:

$$
\hat{F}\phi_i = \epsilon_i \phi_i
$$

görünüşte tek parçacık denklemi gibi ama:

$$
\hat{F} = \hat{F}[\{\phi_j\}]
$$

→ Yani operatör çözümün kendisine bağlıdır.

Bu yüzden:

$$
\text{Nonlinear eigenvalue problem}
$$
