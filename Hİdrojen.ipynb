
Hidrojen Hamiltonyeni (atomik birimlerde):

$$
H = -\frac12 \nabla^2 - \frac1r
$$

S-küresel simetri için:

$$
\langle H \rangle =
\frac{\int_0^\infty \psi(r),H\psi(r)4\pi r^2 dr}
{\int_0^\infty |\psi(r)|^24\pi r^2 dr}
$$
---

# ✅ Kullanılan varyasyon ansatzları

### ⭐ Taban durum (1s)

$$
\psi_{1s}(r,\alpha)=e^{-\alpha r}
$$

Gerçek sonuç → (E=-0.5) Hartree

---

### ⭐ İlk uyarılmış durum (2s)

Ortogonallik için düğüm içeren fonksiyon:

$$
\psi_{2s}(r,\beta)= (1-\beta r),e^{-\beta r}
$$

Gerçek sonuç → (E=-0.125) Hartree

---

#  NÜMERİK ENERJİ HESABI




```python
import numpy as np
from scipy.optimize import minimize_scalar


HARTREE_TO_EV = 27.211386


r_max = 50.0
N = 5000
r = np.linspace(1e-5, r_max, N)
dr = r[1] - r[0]

# -----------------------------
# Laplacian (radial)
# -----------------------------
def radial_laplacian(psi):
    dpsi = np.gradient(psi, dr)
    d2psi = np.gradient(dpsi, dr)
    return d2psi + (2/r)*dpsi

# -----------------------------
# Enerji hesap
# -----------------------------
def energy_ev(psi):
    lap = radial_laplacian(psi)
    Hpsi = -0.5*lap - psi/r
    
    num = np.trapezoid(psi * Hpsi * 4*np.pi*r**2, r)
    den = np.trapezoid(psi**2 * 4*np.pi*r**2, r)
    
    E_hartree = num/den
    return E_hartree * HARTREE_TO_EV

# =============================
# ⭐ 1s
# =============================
def energy_1s(alpha):
    psi = np.exp(-alpha*r)
    return energy_ev(psi)

res1 = minimize_scalar(energy_1s, bounds=(0.1, 2.0), method='bounded')

print("---- 1s sonucu ----")
print("alpha =", res1.x)
print("E_1s =", res1.fun, "eV")

# =============================
# ⭐ 2s
# =============================
def energy_2s(beta):
    psi = (1 - beta*r)*np.exp(-beta*r)
    return energy_ev(psi)

res2 = minimize_scalar(energy_2s, bounds=(0.05, 1.0), method='bounded')

print("\n---- 2s sonucu ----")
print("beta =", res2.x)
print("E_2s =", res2.fun, "eV")
```
