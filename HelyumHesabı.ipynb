```python
import numpy as np
from scipy.optimize import minimize_scalar

HARTREE_TO_EV = 27.211386

r_max = 50.0
N = 5000
r = np.linspace(1e-8, r_max, N)
dr = r[1] - r[0]

def radial_laplacian(psi):
    dpsi = np.gradient(psi, dr)
    d2psi = np.gradient(dpsi, dr)
    return d2psi + (2/r)*dpsi

def helium_energy_ev(alpha):
    # Deneme dalga fonksiyonu: psi(r1, r2) = phi(r1) * phi(r2)
    # Burada phi(r) = exp(-alpha * r)
    psi = np.exp(-alpha * r)
    
    # 1. Kinetik Enerji ve Çekirdek Çekimi 
    lap = radial_laplacian(psi)
    # Helyumda Z=2 olduğu için potansiyel -2/r
    h_core_psi = -0.5 * lap - (2.0 / r) * psi
    
    # Beklenen değer <phi|h_core|phi>
    num_h = np.trapezoid(psi * h_core_psi * 4 * np.pi * r**2, r)
    den = np.trapezoid(psi**2 * 4 * np.pi * r**2, r)
    E_single = num_h / den
    
    # 2. Elektron-Elektron İtmesi (J integrali basitleştirilmiş hali)
    
    V_ee = (5/8) * alpha
    
    # Toplam Enerji: 2*E_tek_elektron + V_ee
    # (Çünkü iki elektron var)
    E_total_hartree = 2 * E_single + V_ee
    
    return E_total_hartree * HARTREE_TO_EV

# Helyum için alpha (etkin çekirdek yükü) genelde 1.6 - 1.8 civarı çıkar
res_he = minimize_scalar(helium_energy_ev, bounds=(1.0, 3.0), method='bounded')

print("---- Helyum (He) Varyasyonel Sonuç ----")
print(f"Optimize edilmiş alpha (Z_eff): {res_he.x:.4f}")
print(f"Hesaplanan Taban Durum Enerjisi: {res_he.fun:.4f} eV")
print(f"Deneysel Değer (Yaklaşık): -79.005 eV")
```
